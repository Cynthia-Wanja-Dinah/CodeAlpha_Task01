#Step 3: EDA(Exploratory Data Analysis)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
# Convert 'States' column to string before plotting
plt.bar(x=df['States'].astype(str), height=df['Estimated Unemployment Rate'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Estimated Unemployment Rate')
plt.title('Estimated Unemployment Rate by State')
plt.tight_layout() # Adjusts subplot params for a tight layout.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'df' is your DataFrame

# Convert the 'Date' column to datetime objects if it's not already
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='Date', y='Estimated Unemployment Rate', hue='Region')
plt.title('Unemployment Rate Over Time (Urban vs Rural)')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('month and Year')

# Format the x-axis to show month and year
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))  # Format to display year and month

plt.grid(True)
plt.show()
```

```python
df.columns
```

```python

plt.figure(figsize=(12, 6))
# Remove the extra space at the end of the column name
sns.boxplot(data=df, x='Region', y='Estimated Unemployment Rate')
plt.title('Distribution of Unemployment Rates (Urban vs Rural)')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = df[['Estimated Unemployment Rate', 'Estimated Labour Participation Rate', 'Estimated Employed']].corr()
# Removed the extra space at the end of 'Estimated Unemployment Rate'

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#swarm plot for unemployement in different   regions
plt.figure(figsize=(10,6))
sns.swarmplot(x='Region',y='Estimated Unemployment Rate',data=df)
```

```python

# Group data by region and calculate the mean of the relevant columns
area_analysis = df.groupby('Region')[['Estimated Unemployment Rate', 'Estimated Labour Participation Rate']].mean()

# Plotting grouped bar plot
area_analysis.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'orange'])
plt.title('Urban vs Rural: Unemployment & Labour Participation Rates')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

#Findings
- The bar chart shows the estimated unemployment rate for each state in India. Some states like Haryana, Tripura, and J&K have relatively higher unemployment rates compared to others.

- from line plot unemployment rates have changed over time, categorized by region (Urban vs. Rural). There are fluctuations in unemployment rates, with some periods showing higher unemployment than others. Rural areas tend to have slightly higher unemployment rates compared to urban areas.

- from the box plot  We can see that the median unemployment rate is higher in rural areas, with a wider range of values.

- The heatmap shows the correlation between 'Estimated Unemployment Rate', 'Estimated Employed', and 'Estimated Labour Participation Rate'. It indicates a strong negative correlation between unemployment rate and employment levels, which is expected. There's also a positive correlation between unemployment rate and labor participation rate, which might suggest that more people entering the workforce could lead to higher unemployment if job creation doesn't keep pace.

- The bar plot clearly shows that urban areas have a significantly higher average estimated unemployment rate compared to rural areas. This might seem counterintuitive, but it could be attributed to factors like higher population density, greater competition for jobs, and a potential influx of job seekers from rural areas to urban centers.

- Conversely, rural areas demonstrate a higher average estimated labor participation rate compared to urban areas. This indicates that a larger proportion of the rural population is actively engaged in the workforce, either employed or seeking employment.
